# Data Preparation

## Inconsistency

```python
print(f" The data has {heart_data.duplicated().sum()} duplicated rows")
```

```python
# droppping the duplicated rows
heart_data = heart_data.drop_duplicates()

# confirming whether the duplicates have been removed
heart_data.duplicated().sum()
```

## Incompleteness

```python
# Checking for missing values
heart_data.isna().sum()
```

## Feature Engineering

```python
# Creating a new column from the BMI column
# 0 = underweight
# 1 = normal
# 2 = overweight
# 3 = obese
heart_data.loc[heart_data['BMI'] < 18.5,'weight'] = 0
heart_data.loc[(heart_data['BMI'] >= 18.5) & (heart_data['BMI'] < 25),'weight'] = 1
heart_data.loc[(heart_data['BMI'] >= 25) & (heart_data['BMI'] < 30),'weight'] = 2
heart_data.loc[(heart_data['BMI'] >= 30),'weight'] = 3
```

```python
# Type casting categorical columns to reflect the categorical nature
cat_col = ['HeartDiseaseorAttack', 'HighBP', 'HighChol', 'CholCheck',
       'Smoker', 'Stroke', 'Diabetes', 'PhysActivity', 'Fruits', 'Veggies',
       'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 'GenHlth', 
       'DiffWalk', 'Sex', 'Age', 'Education','Income', 'weight']
for col in cat_col:
    heart_data[col] = heart_data[col].astype('int64')
    heart_data[col] = heart_data[col].astype('category')
```

## Exploratory Data Analysis

```python
plt.figure(figsize=(15,40))
for i,col in enumerate(cat_col):
    plt.subplot(len(cat_col), 2, i+1)
    plt.suptitle("Plot Value Count", fontsize=20, x=0.5, y=1)
    sns.countplot(data=heart_data, x=col)
    plt.title(f"{col}")
    plt.tight_layout()
plt.show()
```
